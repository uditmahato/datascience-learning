```python
import pandas as pd
df=pd.read_csv('titanic.csv')
df.head()
```

```python
df.shape
```

```python
df.describe()
```

# lets clean our data

```python
da=df[['sex','age','fare','survived']]
da.head()
```

```python
da.describe()
```

```python
da.dropna()
```

```python
x=da.drop(columns='survived')
```

```python
y=da['survived']
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
model.Fit(x,y)
model.predict(['male',17,157.5])
```
