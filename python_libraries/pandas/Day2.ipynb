```python
data=[[1,65],[2,45],[3,56],[4,87],[5,77],[6,98],[7,56],[8,83],[9,45],[10,89]]
index=['Ram','Shyam','Hari','Krishna','Falao','chilano','manisha','aryal','chaudhary','ramu']
li1=pd.DataFrame(data,index=index,columns=['Test','Marks'])
li1
```

```python
test
```

```python
li2=test.merge(li1)
li2
```

```python
# left outer join(left column stays complete)
test2=test.merge(li1,how='left')
test2
```

```python
# right outer join(right column stays complete)
test3=test.merge(li1,how='right')
test3
```

```python
#preserve all values(outer join)
test3=test.merge(li1,how='outer')
test3
```

```python
#merge rows by index
test4=test.merge(li1,left_index=True,right_index=True)
test4
```
