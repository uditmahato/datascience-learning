# Working with Series
Series is a 1d ,homogeniously typed array

```python
s=pd.Series([15,16,17,22,7],name="ages")
s
```

Lets put index in the data series

```python
s=pd.Series([15,16,17,22,7],index=["a","b","c","d","e"],name="ages")
s
```
