```python
import pandas as pd
```

```python
df=pd.read_csv('baseball.csv')
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
#start new diagram witha figure
plt.figure()
```

# Scatter plot

```python
#draw scatter plot
df.plot.scatter('Age','Height')
```

# bar plot

```python
df.plot.bar(x='Age',y='Weight',width=0.7)
```

# area plot


```python
df.plot.area(stacked=True,alpha=1.0)
```

# Box plot

```python
df.plot.box()
```

# Histogram over 1 columns

```python
#over one columns
df['Age'].plot.hist(bins=1)
```

```python
#over multiple columns
df.plot.hist(bins=10,alpha=0.5)
```

```python
#set tick marks
labels=['A','B','C']
```

```python
#Label diagram
plt.title('relation between age and height')
plt.xlabel('Age of players')
plt.ylabel('Height of players')
df.plot.scatter('Age','Height')
```
